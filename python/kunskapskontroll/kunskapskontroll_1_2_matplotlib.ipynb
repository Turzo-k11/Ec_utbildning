### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Plotting in MAtplotlib can be done in either OO(Object-Oriented approach) or with pyplot/MATLAB style. 
In general it's recomended to use OO style. 
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
Figure: This is the whole figur and can be seen as a canvas.  
Axes: The part of the "canvas" that the plot is attached on. 
Axis: Sets the scale and limit both horizontally(x) and vertically(y).
An artist: Everything visible on the figure is an artist.
```

### When plotting in Matplotlib, what is the expected input data type?

```python
numpy.array or objects that can be passed to numpy.asarray.
Classes that are similar to arrays such as pandas data objects.
Common convention is to convert these to numpy.array objects prior to plotting.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4, 4, 100)
y = x**2

fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title("y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
#Pyplot approach
plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())


#Subplots
fig, ax = plt.subplots(1, 2, layout = "constrained")
fig.suptitle("2 Different Plots", size = 25)

#Scatter plot
ax[0].scatter(random_data_x, random_data_y, color= 'r')
ax[0].set_title('Scatter Plot')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

#Bar plot
ax[1].bar(names, values, color= 'b')
ax[1].set_title('Bar Plot')
ax[1].set_xlabel('Categories')
ax[1].set_ylabel('Values')
```
